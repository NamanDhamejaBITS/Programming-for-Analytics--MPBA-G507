## Merging & joining
* Concatenate using concat() method
* Joins: left, right, inner, & outer

```python
# load the numpy library with alias name "np"
import numpy as np
# load the pandas library with alias name "pd"
import pandas as pd
```

```python
# Create p2 & p1 dataframe that constitutes column names:
# Students, Group, Gender, Scores, Student ID
p2_df = pd.DataFrame(
 {
 "Students": [
 "Rohit, Mr. Rohit",
 "Jyothula, Mr. Sougandh",
 "Mary, Miss. Sajitha",
 "Vaibhav, Mr. Vaibhav"
 ],
"Group": [ "P2", "P2", "P2",  "P2"],
"Gender" : ["Male", "Male", "Female", "Male" ],
"Scores" : [8, 9, 7, 8 ],
"Student_ID" : ["2021MBA01","2021MBA02","2021MBA03", "2021MBA05"]
 }
)

# view p2
p2_df
```

```python
p1_df = pd.DataFrame(
 {
 "Students": [
  "J, Miss. Nehal",
  "Priyanka, Miss. Mira"
 ],
"Group": [ "P1", "P1"],
"Gender" : [ "Female", "Female"],
"Scores" : [7,  8],
"Student_ID" : ["2021MBA04",  "2021MBA06"]
 }
)

# view p1
p1_df
```

```python
# Concatenate using concat() method into class_df
class_df = pd.concat(objs = [p2_df, p1_df])
class_df
```

```python
# joins
band_members = pd.DataFrame({"name": ["John", "Paul", "Mick"], 
"band": ["Beatles", "Beatles", "Stones"], "likes": ["food", "baseball", "Texas"]})
print("band_members :\n", band_members)

band_instruments =pd.DataFrame({"name": ["John", "Paul", "Keith"], 
 "plays" : ["guitar", "bass", "guitar"], "food": ["Mexican", "Italian", "Pizza"]})

print("band_instruments : \n", band_instruments)
```

```python
# left join
band_members.merge(band_instruments, how = "left", on = "name")
```

```python
# right join
band_members.merge(band_instruments, how = "right", on = "name")
```

```python
# inner join
band_members.merge(band_instruments, how = "inner", on = "name")
```

```python
# outer join
band_members.merge(band_instruments, how = "outer", on = "name")
```

```python
# outer join
band_members.merge(band_instruments, how = "outer",  on = "name",  indicator = True)
```

```python
# outer join
band_members.merge(band_instruments, how = "outer",  left_on = "name", right_on = "name", indicator = True)
```
